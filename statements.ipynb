### Control Structures

##### For Loop

```python
fruits = ["apple", "pear", "avacado"]
for fruit in fruits:
    print(f"A fruit that I enjoy often is a {fruit}")
```

```python
for fruit in fruits:
    if fruit == "apple":
        print(f"I love this fruit {fruit}")
        break
```
